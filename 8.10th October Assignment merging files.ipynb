```python
from glob import glob
import pandas as pd
```

```python
all_files = glob('/Users/ashish/Downloads/AReM/walking/dataset*.csv')
```

```python
all_files
```

```python
li = []
```

```python
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, skiprows = 4)
    li.append(df)
```

```python
print(li)
```

```python
df_merger = pd.concat(li, ignore_index=True, axis=0)
```

```python
df_merger.to_csv('/Users/ashish/Downloads/AReM/walking/finalemerger.csv')
```
